*Importing the libraries*

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to pandas dataframe
df = pd.read_csv("/content/creditcard.csv")
```

```python
# printing first 5 rows of the given dataset
df.head()
```

```python
# printing last 5 rows of the given dataset
df.tail()
```

```python
# dataset information
df.info()
```

```python
# checking no. of missing values in each column
df.isnull().sum()
```

```python
# distribution of legit and fradulent transactions
df['Class'].value_counts()
```

The given dataset is highly unbalanced


```python
# '0' --> indicates normal or legit transaction
# '1' --> indicates fradulent transaction
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing the values for both transactions
df.groupby('Class').mean()
```

Constructing a similar dataset containing a similar distribution of legit and fradulent transactions

Number of fradulent transactions --> 52

```python
legit_samp = legit.sample(n=52)
```

```python
# Concatenating 2 dataframes
new_df = pd.concat([legit_samp,fraud],axis = 0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

Now this is a uniformly distributed dataset

```python
new_df.groupby('Class').mean()
```

Hence the nature of the dataset is not changed as the mean values are same

**Splitting the dataset into features and targets**

```python
x = new_df.drop(columns = 'Class',axis = 1)
y = new_df['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting the data into training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,stratify = y,random_state = 2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# ***Training the model using Logistic Regression***

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

# Model Evaluation

```python
# accuracy on train data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy score of training data: ",training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy score of test data: ",test_data_accuracy)
```
